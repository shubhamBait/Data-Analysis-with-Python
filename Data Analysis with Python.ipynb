```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv")
data.head()
```

```python
data.drop(['id', 'Unnamed: 0'], axis = 1, inplace = True)
data.describe()
```

```python
data.isnull().sum()
```

```python
mean=data['bedrooms'].mean()
data['bedrooms'].replace(np.nan,mean, inplace=True)
```

```python
mean=data['bathrooms'].mean()
data['bathrooms'].replace(np.nan,mean, inplace=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
x = data[['sqft_living']]
y = data.price
lr = LinearRegression()
lr.fit(x, y)
lr.score(x, y)
```

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
X = data[features]
y = data.price
lr.fit(X, y)
lr.score(X, y)
```

```python
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
X = data[features]
y = data.price
pipe = Pipeline(Input)
pipe.fit(X, y)
pipe.score(X, y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = data[features]
Y = data['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
rm = Ridge(alpha=0.1)
rm.fit(x_train, y_train)
rm.score(x_test, y_test)
```

```python
pr = PolynomialFeatures(degree = 2)
X_train_pr = pr.fit_transform(x_train)
X_test_pr = pr.fit_transform(x_test)

rr = Ridge(alpha = 0.1)
rr.fit(X_train_pr, y_train)
rr.score(X_test_pr, y_test)
```
